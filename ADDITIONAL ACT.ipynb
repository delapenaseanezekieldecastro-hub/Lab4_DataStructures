```python
from collections import deque, Counter

CONTROL_NUM = 5
LOCAL_HAZARD = "FLOOD"

# Fixed window size
dq = deque(maxlen=5)

# Alert stream
alerts = [
    (5, "WEATHER"),
    (6, "TRAFFIC"),
    (7, "WEATHER"),
    (8, "WEATHER"),
    (9, LOCAL_HAZARD),
    (10, "WEATHER"),
    (11, "TRAFFIC")
]

alerts.extend([
    (12, "WEATHER"),
    (13, "TRAFFIC"),
    (14, "WEATHER"),
    (15, LOCAL_HAZARD),
    (16, "TRAFFIC"),
    (17, "WEATHER"),
    (18, "WEATHER")
])

# Process stream
for alert in alerts:
    categories = [x[1] for x in dq]
    count = Counter(categories)

    # allow max 2 per category
    if count[alert[1]] < 2:
        dq.append(alert)

print("Filtered Alerts:")
print(dq)

# Sum of IDs
total = sum(x[0] for x in dq)
print("Total ID Sum:", total)
```

```python
CONTROL_NUM = 5
FAVORITE_ARTIST = "SB19"

# Festival lineup
festival = {
    "Ben&Ben", "SB19", "BINI", "Eraserheads",
    FAVORITE_ARTIST, "Zild", f"Indie {CONTROL_NUM}"
}

festival.update({
    "Moira", "December Avenue", "IV of Spades",
    "Unique", "Cup of Joe"
})

# User preferences
user_a = {"Ben&Ben", "BINI", "Maki", "Dionela", FAVORITE_ARTIST}
user_b = {"SB19", "Eraserheads", "Zild", f"Indie {CONTROL_NUM}", "Parokya"}

# Jaccard similarity
def jaccard(a, b):
    return len(a & b) / len(a | b) * 100

print("User A similarity:", jaccard(user_a, festival))
print("User B similarity:", jaccard(user_b, festival))

# Missing artists
missing = user_a - festival
print("User A missing artists:", missing)
```
